Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
# -- ВАШ КОД ТУТ --
# -- ВАШ КОД ТУТ --
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
# данные получены с помощью функции make_classification 
with open('clustering.pkl', 'rb') as f:
    data_clustering = pickle.load(f)

X = np.array(data_clustering['X'])
Y = np.array(data_clustering['Y'])


model = DBSCAN(min_samples=9)
model.fit(X)

plt.scatter(X[:, 0], X[:, 1], s=40, c=model.labels_, marker='o', alpha=0.8, label='data')
plt.show()
```
