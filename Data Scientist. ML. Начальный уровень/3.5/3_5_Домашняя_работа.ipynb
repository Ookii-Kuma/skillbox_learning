Есть два набора точек - *x_hw* и *y_hw*. В рамках домашней работы нужно
* визуализировать набор точек
* найти коэффициенты регрессии $w_0, w_1$ по шагам, как в уроке
* посчитать предсказание в виде $\hat{y} = w_0 + w_1x$ и визуализировать его вместе с точками *x_hw* и *y_hw*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv

plt.style.use('ggplot')
```

```python
def lin_reg_coef(X,y):
    return inv((X.T).dot(X)).dot(X.T).dot(y)

def lin_reg(coef, X):
    return coef[0] + coef[1]*X
```

```python
x_hw = [50, 60, 70, 100]
y_hw = [10, 15, 40, 45]
```

```python
df = pd.DataFrame(zip(x_hw, y_hw), columns = ['X', 'y'])
```

```python
df.plot(x='X', y='y')
plt.title('График переменной y')
```

```python
coef = lin_reg_coef(np.array([list(pair) for pair in zip([1]*len(x_hw), x_hw)]), df['y'].values.reshape(-1,1))
df['y_pred'] = lin_reg(coef, x_hw)
```

```python
df.plot(x='X')
plt.title('График линейной регрессии')
```
